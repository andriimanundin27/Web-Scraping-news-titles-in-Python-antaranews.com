```
from google.colab import drive
drive.mount('/content/drive')
%cd /content/drive/'My Drive'/'data'/

import requests 
from bs4 import BeautifulSoup 
import pandas as pd
```

```
page = 1
judulss=[]
while (page <= 20): 
  print(page)
  url ="https://www.antaranews.com/nasional/umum/"+str(page) 
  film = requests.get(url) 
  page=int(page) 
  page += 1 #print(url) 
  soup=BeautifulSoup(film.content,'html.parser')
  judul = soup.select("h3 a")
  juduls = [js.get_text() for js in judul]
  judulss+=juduls[2:]
```

```
df = pd.DataFrame({'col':judulss})
# df.to_csv('kalimat deklaratif1000.csv')
```
